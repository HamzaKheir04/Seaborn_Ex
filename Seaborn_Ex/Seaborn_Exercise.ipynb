# EX 1 - Distribution Plots

Create a histogram of a given dataset using `histplot`.

```python
Titanic = sns.load_dataset("titanic")
Titanic.head()
```

```python
sns.histplot(Titanic["age"])
```

Generate a KDE plot of the same dataset using `kdeplot`.

```python
sns.kdeplot(Titanic["age"])
```

Use `jointplot` to display the relationship between two numerical variables.

```python
sns.jointplot(x="age",y="fare",data=Titanic,kind="hex",color="royalblue")
```

Create a `pairplot` to visualize pairwise relationships in a dataset.

```python
sns.pairplot(Titanic,hue="class" ,palette="viridis")
```

```python
Titanic.head()
```


## EX 2 - Categorical Data Plots
Create a `barplot` to show the average values of a categorical variable.

```python
sns.barplot(Titanic,x="sex",y="fare")
```

Generate a `countplot` to show the frequency of different categories.

```python
sns.countplot(Titanic,x="class")
```

Use `boxplot` to visualize the distribution of numerical data across categories.

```python
sns.boxplot(Titanic,x="who",y="age")
```

```python
Titanic.head()
```

Create a `violinplot` to combine KDE and boxplot visualizations.

```python
sns.violinplot(x="sex",y="age",data=Titanic)
sns.kdeplot(x="age",data=Titanic)
sns.boxplot(x="sex",y="age",data=Titanic)
```

Display individual data points in a category using `stripplot`.

```python
sns.stripplot(Titanic,x="class",y="age")
```

Use `swarmplot` to create a more readable version of `stripplot`.

```python
sns.swarmplot(Titanic,x="class",y="age",s=4)
```

# EX 3 - Matrix Plots
Create a heatmap to display correlations in a dataset using `heatmap`.

```python
Corr = Titanic[["age","fare","pclass"]].corr()
Corr
```

```python
sns.heatmap(Corr)
```

```python
Titanic.head(10)
```

Use `clustermap` to cluster data based on similarity.

```python
Pivot = Titanic.pivot_table(values="age",index="class",columns = "sex",observed=False)
sns.heatmap(Pivot)
```

# EX 4 - Grids
Implement a `PairGrid` to customize pairwise plots.

```python
G = sns.PairGrid(Titanic)
G.map_upper(sns.scatterplot, color='#034078')      # Upper triangle: scatter
G.map_diag(sns.histplot,color='#001f54')            # Diagonal: histogram with KDE
G.map_lower(sns.kdeplot)                           # Lower triangle: KDE
```

Use `FacetGrid` to create multiple subplots based on a categorical variable.

```python
g1 = sns.FacetGrid(Titanic , col = "sex", row="class")
g1 = g1.map(plt.hist,"age")
```

```python
Titanic.head()
```

Implement a `JointGrid` to create a customized joint distribution plot.

```python
g2 = sns.JointGrid(x="fare", y="age", data=Titanic)
g2 = g2.plot(sns.regplot, sns.distplot) # safe to ignore the Warning
```

# EX 5 - Style and Customization
Change the Seaborn style using `set_style`.

```python
sns.set_style("darkgrid")
sns.histplot(Titanic["who"])
```

Remove spines from a plot using `despine`.

```python
sns.countplot(x='who',data=Titanic)
sns.despine(offset={"top":2,"bottom":5,"right":6,"left":4},top=False,right=False)
```

Adjust figure size and aspect ratio using `set_context`.

```python
sns.set_context("notebook", font_scale=1.25)
sns.countplot(x='sex', data=Titanic,palette="Paired")
```

Customize color palettes using `set_palette`.

```python
sns.set_palette("pastel")
sns.countplot(x='sex', data=Titanic)
```
